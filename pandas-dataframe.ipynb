**Handling Null Values**

```python
stud= {"marks":[23,None,56,None],    
        "name":['raha','jenna','joe',None],
     "age":[56,None,23,None]
      }
df=pd.DataFrame(stud)
df
```

```python
df.dropna()
```

```python
df.dropna(how='all',inplace=True)
df                                 #how='all',it drop only those records where every cell has a null value
                                  #inplace= True make the permanent changes in a dataframe
```

```python
#axis
df.dropna(axis=1)  #axis=0 perform row wise and axis=1 perform column wise
```

```python
#subset
df
```

```python
df.dropna(axis=0,subset=['age'])
```

**Ranking**

```python
d={
    'Country':['India','China','Japan','Bangaladesh','Singapore'],
    'Gold':[12,20,11,3,None]
}
a=pd.DataFrame(d)
a
```

```python
a['Rank']= a['Gold'].rank()
a
```

```python
a['Rank']= a['Gold'].rank(ascending=False)
a.sort_values('Rank',inplace=True)
a
```

```python
a['Rank']= a['Gold'].rank(ascending=False,na_option='bottom')
a
```

**Find duplicates**

```python
stud={
    'name':['joe','mon','joe','ross','phoebe','joe'],
    'marks':[43,21,45,46,67,43],
    'age':[22,30,25,34,56,22]
}
df=pd.DataFrame(stud)
df
```

```python
n=df['name'].duplicated()
df[n]
```

```python
n1=df['name'].duplicated(keep='last')  #it preserves the last occurence of duplicated record 
df[n1]
```

```python
n2=df['name'].duplicated(keep=False) #it fetches all duplicated the record
df[n2]
```

```python
n3=df.duplicated(subset=['name','marks','age'])
df[n3]
```
